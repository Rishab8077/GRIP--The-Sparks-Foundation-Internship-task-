```python
# importing library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
numeric_feat = [feature for feature in dataset.columns if dataset[feature].dtype !='O']
print('No. of numeric feature is',len(numeric_feat))
dataset[numeric_feat].head
```

```python
for feat in numeric_feat:
    print(feat + "=",len(dataset[feat].unique()))
```

```python
discrete_feat=['Quantity','Discount']
```

```python
    dataset.groupby('Quantity')['Profit'].median().plot.bar()
    plt.xlabel('Quantity')
    plt.ylabel('Profit')
    plt.show
```

```python
    dataset.groupby('Discount')['Profit'].median().plot.bar()
    plt.xlabel('Discount')
    plt.ylabel('Profit')
    plt.show
```

### We can observe how profit is affected by Discount and Quantity

```python
for feature in numeric_feat:
    data=dataset.copy()
    sns.displot(data[feature],kde=True,bins=10)
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.show()
```

```python
#dropping postal code
dataset.drop('Postal Code',axis=1,inplace=True)
#plotting correlation heatmap chart
corr = dataset.corr()
sns.heatmap(corr,annot=True)
```
