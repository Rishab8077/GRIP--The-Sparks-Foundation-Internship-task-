## Venue where most of the matches were played

```python
winning_team=[]
for i in (match_df['season'].unique()):
    winning_team.append(match_df[match_df['season']==i]['winner'].tail(1).values[0])
print(winning_team)
```

```python
win_team=pd.DataFrame(winning_team,columns=['Teams'])
win_team
```

```python
plt.figure(figsize=(15,8))
sns.countplot('Teams',data=win_team,order=win_team['Teams'].value_counts().index)
plt.title('Winners of IPL across 11 seasons')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot('toss_decision',data=match_df)
plt.show()
```

## Maximum Field was choosed after winning the toss

```python
plt.figure(figsize=(15,8))
sns.countplot('toss_winner',data=match_df,order=match_df['toss_winner'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
match_df['tosswin_gamewin']=np.where((match_df['toss_winner']==match_df['winner']),'Yes','No')
plt.figure(figsize=(15,8))
sns.countplot('tosswin_gamewin',data=match_df,hue='toss_decision')
plt.show()
```
